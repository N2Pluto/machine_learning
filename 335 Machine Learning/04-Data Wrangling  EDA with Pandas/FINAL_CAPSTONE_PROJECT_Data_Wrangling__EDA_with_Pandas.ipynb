```python
#5. Scale the raw Age column (in months) using Normalization.
Kyphosis_df["Age"] = Kyphosis_df["Age"].astype("float64")

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Kyphosis_df['Age'] =scaler.fit_transform(Kyphosis_df['Age'].values.reshape(-1,1))
```

```python
Kyphosis_df.describe().round(2)
```

```python
#5. Scale the raw Age column (in months) using standardization.
Kyphosis_df["Age"] = Kyphosis_df["Age"].astype("float64")

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Kyphosis_df['Age'] =scaler.fit_transform(Kyphosis_df['Age'].values.reshape(-1,1))
```
